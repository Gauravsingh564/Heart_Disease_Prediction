Spliting features into Dependent and independent features

```python
X=new_data.drop(columns='target',axis=1)
y=new_data['target']
```

```python
X
```

```python
y
```

Train_test_split

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
X_train.shape,X_test.shape,X.shape
```

Standardization

```python
scaler=StandardScaler()
```

```python
scaler.fit_transform(X_train)
scaler.transform(X_test)
```

Model training

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

Model Evaluation

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,y_train)
```

```python
print("Accuracy on training data : ",training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,y_test)
```

```python
print("Accuracy on testing data : ",testing_data_accuracy)
```

Building a Predictive System

```python
input_data=[57,1,0,130,131,0,1,115,1,1.2,1,1,3]

#changing data into numpy array

intput_data_as_nparray=np.asarray(input_data)

#reshape the numpy array as we are predicting only for one instaance

input_data_reshaped=intput_data_as_nparray.reshape(1,-1)

prediction=model.predict(input_data_reshaped)

print(prediction[0])

if (prediction[0]==1):
    print("Person has heart disease")
else:
    print("Person does not have heart disease")
```
